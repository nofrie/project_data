```python
import pandas as pd
import os

pwd= os.getcwd()
# Replace 'your_file.xlsx' with the actual name of your Excel file
file_path = pwd + '/Downloads/109-111年度產品類別銷售分析表.xlsx'

# Read the Excel file into a pandas DataFrame
df = pd.read_excel(file_path)

# Display the DataFrame
df.head()
```

```python
# Filter and select columns for '109年'
df_109 = df.iloc[:, [0, 1, 2, 3, 4, 5, 6]]

# Filter and select columns for '110年'
df_110 = df.iloc[:, [0, 1, 7, 8, 9, 10, 11]]

# Filter and select columns for '111年'
df_111 = df.iloc[:, [0, 1, 12, 13, 14, 15, 16]]

# Set '期間：1 / 1- 12 / 31', 'Unnamed: 1' as the index for all three DataFrames
df_109.set_index(['期間：1 / 1- 12 / 31', 'Unnamed: 1'], inplace=True)
df_110.set_index(['期間：1 / 1- 12 / 31', 'Unnamed: 1'], inplace=True)
df_111.set_index(['期間：1 / 1- 12 / 31', 'Unnamed: 1'], inplace=True)

df_110.columns
```

```python
# List of DataFrames
dataframes = [df_109, df_110, df_111]

# Loop through each DataFrame
for df in dataframes:
    # Set new header
    df.columns = df.iloc[0].tolist()
    
    # Set new index name
    df.index.names = tuple(df.index[0])
    
    # Exclude the first row after setting header and index name
    df = df[1:]
```

```python
# Define a function for grouping
def group_and_aggregate(df):
    # Set new header
    df.columns = df.iloc[0].tolist()
    
    # Set new index name
    df.index.names = tuple(df.index[0])
    
    # Exclude the first row after setting header and index name
    df = df[1:]

    # Extract the first letter of '類別' index values
    first_letters = df.index.get_level_values('類別').str[0]

    # Create a new index using the first letter
    new_index = pd.Index(first_letters, name='Grouped_類別')

    # Group by both the new index and '類別名稱', then aggregate
    grouped_df = df.groupby([new_index, '類別名稱']).agg(lambda x: x[0])

    return grouped_df
    
grouped_dfs = {}

# Loop through each DataFrame
for i, df in enumerate(dataframes, start=109):
    grouped_dfs[f'df_{i}_grouped'] = group_and_aggregate(df)

# Display or further process the grouped DataFrames
for key, value in grouped_dfs.items():
    print(f'{key}:\n{value}\n')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Draw a figure with three subplots
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Iterate through grouped DataFrames
for ax, (key, value) in zip(axes, grouped_dfs.items()):
    # Calculate aggregated data
    agg_data = value.groupby('Grouped_類別').sum()

    # Sort values based on '銷售金額' descending
    top5_data = agg_data.sort_values(by='銷售金額', ascending=False).head(5)

    # Scatter plot for the top 5
    sns.scatterplot(data=top5_data, x='銷售金額', y='銷售利潤', hue=top5_data.index, palette='viridis', ax=ax)\
        .set(title=f'{key.split("_")[1]}年銷售金額top5')

    # Annotate the hue next to each point
    for index, row in top5_data.iterrows():
        ax.text(row['銷售金額'], row['銷售利潤'], index, ha='center', va='bottom')

# Set the font
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']

plt.tight_layout()
# Show the plots
plt.show()
```

```python
# Create subplots
fig, axes = plt.subplots(3, 2, figsize=(9, 9))

# Loop through grouped_dfs for the first plot (bar plot)
for ax_row, (key, value) in zip(axes, grouped_dfs.items()):
    # Group by '類別名稱' and sum
    agg_data = value.groupby('類別名稱').sum()

    # Sort values based on '銷售金額' descending
    top5_categories = agg_data.sort_values(by='銷售金額', ascending=False).head(5)

    # Bar plot for the top 5
    sns.barplot(ax=ax_row[0], data=top5_categories, x=top5_categories.index, y='銷售金額')
    
    # Set subplot title
    ax_row[0].set_title(f'{key.split("_")[1]}年top5銷售品項ㄎㄎ')

    # Scatter plot for the top 5
    sns.scatterplot(ax=ax_row[1], data=top5_categories, x='銷售金額', y='銷售利潤', hue=top5_categories.index)
    
    # Set subplot title
    ax_row[1].set_title(f'{key.split("_")[1]}年top5金額/利潤')

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```
